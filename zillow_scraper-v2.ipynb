```python
from splinter import Browser
from bs4 import BeautifulSoup as bsoup
from urllib.request import urlopen
import requests
from splinter import Browser
import pandas as pd
import regex as re
import time
```

```python
pages = []
price_all_data = []
address_all_data = []
stats_all_data = []
types_all_data = []

price_all_data.clear()
address_all_data.clear()
stats_all_data.clear()
types_all_data.clear()

for i in range(1, 4):
    url = str('https://www.zillow.com/new-york-ny/'+str(i)+'_p/')
    pages.append(url)

    executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
    browser = Browser('chrome', **executable_path, headless=False)
    browser.visit(url)
    time.sleep(3)
    html = browser.html
    soup = bsoup(html, 'html.parser')
    
    price = soup.find_all('div', class_='list-card-price')

    address = soup.find_all(class_='list-card-addr')
    
    stats = soup.find(class_='list-card-details')

    types = soup.find_all(class_='list-card-type')

    for a in price:    
        price_all_data.append(a)
    for b in address:    
        address_all_data.append(b)
    for c in stats:
        stats_all_data.append(c)
    for d in types:
        types_all_data.append(d)
        
    price.clear()   
    address.clear()
    stats.clear()  
    types.clear()
    browser.quit()  
 
```
